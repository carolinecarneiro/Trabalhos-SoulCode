# Dicionário dos dados

**olist_marketing_qualified_leads_dataset**

Sobre: Dados de leads qualificados.

* mql_id: ID de lead qualificado

* first_contact_date: Data da primeira solicitação de contato com o representante de vendas

* landing_page_id: ID da landing page onde o lead foi adquirido

* origin: Tipo de mídia de onde o lead foi adquirido (busca paga, busca orgânica, rede social, etc.)

**olist_closed_deals_dataset**

Sobre: Dados de leads que realizaram consulta com o representante de vendas e finalizaram o cadastro na plataforma, tornando-se vendedores.

* mql_id: ID de lead qualificado

* seller_id: ID de vendedor

* sdr_id: ID do Representante de Desenvolvimento de Vendas

* sr_id: ID do representante de vendas

* won_date: Data em que o cadastro do vendedor foi finalizado

* business_segment: Segmento de negócio do lead

* lead_type: Tipo do lead

* lead_behavior: Perfil comportamental do lead identificado pelo Representante de Desenvolvimento de Vendas. A classificação é baseada na metodologia DISC, usando nomes de animais para se referir aos padrões comportamentais

* has_company: Informa se o lead tem CNPJ
has_gtin: Informa se o lead tem GTIN (Global Trade Item Number ou Número Global de Item Comercial)
* average_stock: Estoque médio informado pelo lead

* business_type: Tipo do negócio do lead (revenda/manufatura/etc.)

* declared_product_catalog_size: Tamanho do catálogo declarado do lead

* eclared_monthly_revenue: Receita mensal estimada do lead

**olist_customers_dataset**

* Sobre: Dados dos clientes.

* customer_id: ID do cliente. Note que cada pedido é associado a um customer_id diferente, de modo que um mesmo cliente pode estar associado a vários customer_id

* customer_unique_id: ID único do cliente. Ao contrário do customer_id, há apenas um customer_unique_id por cliente

* customer_zip_code_prefix: Primeiros 5 dígitos do CEP do cliente

* customer_city: Cidade do cliente

* customer_state: Estado do cliente

**olist_geolocation_dataset**

Sobre: Tabela que associa cada prefixo de CEP a uma latitude, longitude, cidade e estado.

* geolocation_zip_code_prefix: Primeiros 5 dígitos do CEP

* geolocation_lat: Latitude

* geolocation_lng: Longitude

* geolocation_city: Cidade

* geolocation_state: Estado

**olist_order_items_dataset**

Sobre: Tabela contendo os dados de cada item no pedido.

* order_id: ID do pedido. Note que cada item no pedido gera um novo registro com o mesmo order_id e outro order_item_id

* order_item_id: ID sequencial indicando um item no pedido. O primeiro item do pedido tem order_item_id 1, o segundo item do pedido tem order_item_id 2, etc.

* product_id: ID do produto

* seller_id: ID do vendedor

* shipping_limit_date: Prazo para que o vendedor envie o pedido ao parceiro de logística

* price: Preço do item

* freight_value: Valor do frete do item

**olist_order_payments_dataset**

Sobre: Dados de pagamento do pedido.

* order_id: ID do pedido. Note que um cliente pode pagar o mesmo pedido de diferentes formas, de modo que cada tipo de pagamento gera um novo registro com o mesmo order_id e outro payment_sequential

* payment_sequential: ID sequencial para cada tipo de pagamento. Por exemplo, se um cliente paga o pedido por voucher e cartão de crédito, o pagamento por voucher recebe o valor de payment_sequential 1 e o pagamento por cartão de crédito recebe o valor de payment_sequential 2

* payment_type: Tipo de pagamento (cartão de crédito, boleto, voucher, etc.)

* payment_installments: Parcelas do pagamento

* payment_value: Valor total do pagamento (preço + frete)

**olist_order_reviews_dataset**

Sobre: Avaliações de clientes após a entrega do pedido através de questionários de satisfação.

* review_id: ID da avaliação

* order_id: ID do pedido

* review_score: Nota de 1 a 5

* review_comment_title: Título da avaliação

* review_comment_message: Mensagem da avaliação

* review_creation_date: Data em que o questionário de satisfação foi enviado ao cliente

* review_answer_timestamp: Data em que o cliente respondeu o questionário de satisfação

**olist_orders_dataset**

Sobre: Dados de pedidos.

* order_id: ID do pedido

* customer_id: ID do cliente

* order_status: Status do pedido (entregue, enviado, cancelado, etc.)

* order_purchase_timestamp: Informa quando o pedido foi realizado

* order_approved_at: Informa quando o pagamento do pedido foi aprovado

* order_delivered_carrier_date: Informa quando o pedido foi enviado ao parceiro logístico

* order_delivered_customer_date: Informa quando o cliente recebeu o pedido

* order_estimated_delivery_date: Informa o prazo estimado de recebimento do pedido anunciado ao cliente no momento da compra

**olist_products_dataset**

Sobre: Dados de produtos.

* product_id: ID do produto

* product_category_name: Categoria do produto

* product_name_lenght: Quantidade de caracteres do nome do produto

* product_description_lenght: Quantidade de caracteres da descrição do produto

* product_photos_qty: Quantidade de fotos publicadas do produto

* product_weight_g: Peso do produto, em gramas

* product_length_cm: Comprimento do produto, em centímetros

* product_height_cm: Altura do produto, em centímetros

* product_width_cm: Largura do produto, em centímetros

**olist_sellers_dataset**

Sobre: Dados dos vendedores.

* seller_id: ID do vendedor
* seller_zip_code_prefix: Primeiros 5 dígitos do CEP do vendedor
* seller_city: Cidade do vendedor
* seller_state: Estado do vendedor

**product_category_name_translation**

Sobre: Tabela de tradução das categorias em product_category_name do português ao inglês.

* product_category_name: Nome da categoria em português
* product_category_name_english: Nome da categoria em inglês

**setores_varejo**

Sobre: Tabela com dados compilados pelo grupo contendo informações de faturamento e crescimento de setores do varejo.

* setor: Nome do setor
* 2016: Faturamento do setor em 2016, em bilhões de reais
* 2017: Faturamento do setor em 2017, em bilhões de reais
* 2018: Faturamento do setor em 2018, em bilhões de reais
* Total: Faturamento do setor de 2006 a 2018, em bilhões de reais
* RANKING: Classificação do setor em ordem crescente em relação ao faturamento total (o setor com o menor faturamento total recebe o valor 1 e o setor com o maior faturamento total recebe o valor 16)
* % Crescimento: Crescimento do setor de 2016 a 2018, em porcentagem
* % Ponderado: Crescimento do setor de 2016 a 2018, em porcentagem, ajustado de acordo com o faturamento total: crescimentos referentes a faturamentos no primeiro quartil (rankings 1-4) são multiplicados por 0.25, crescimentos referentes a faturamentos no segundo quartil (rankings 5-8) são multiplicados por 0.5, crescimentos referentes a faturamentos no terceiro quartil (rankings 9-12) são multiplicados por 0.75 e crescimentos referentes a faturamentos no quarto quartil (rankings 13-16) são multiplicados por 1
* CRESCIMENTO: Nível do crescimento (ALTO e BAIXO) classificado de acordo com a mediana do crescimento ponderado: 50% dos setores com percentuais de crescimento ponderado mais baixos são considerados de baixo crescimento e os 50% setores com percentuais de crescimento ponderados mais altos são considerados de alto crescimento
* PARTICIPAÇÃO: Nível de participação relativa no mercado do setor (ALTO e BAIXO) classificado de acordo com a mediana do nível de faturamento do setor na Olist: 50% dos setores com faturamentos mais baixos são considerados de baixa participação e os 50% setores com faturamentos mais altos são considerados de alta participação

## Criando DataFrames com ticket médio e LTV (lifetime value) por setor

Definimos ticket médio pelo total de faturamento dividido pelo total de pedidos, obtendo uma média do quanto o setor fatura por pedido.

LTV (lifetime value) representa a receita gerada por um consumidor ao longo do período em que mantém um relacionamento com a marca. É calculado pelo ticket médio multiplicado pelo tempo médio de retenção do cliente e pelo número médio de compras realizadas por cliente. No nosso caso, não temos dados sobre o tempo médio de retenção, então consideramos apenas o ticket médio e o número médio de compras.

Fonte: https://pagar.me/blog/ltv-lifetime-value/

```python
# Fazer contagem de pedidos e clientes por setor
df_setor_pedidos = df_merge2[['setor_varejo', 'id_pedido', 'id_cliente_unica']].groupby('setor_varejo').nunique().reset_index()
df_setor_pedidos.rename(columns={'id_pedido': 'qtd_pedidos', 'id_cliente_unica': 'qtd_clientes'}, inplace=True)

# Mesclar o DataFrame de setores para obter o faturamento
df_setor_pedidos = df_setor_pedidos.merge(df_fat_setores, left_on='setor_varejo', right_on='setor_varejo')

# Calcular ticket médio e LTV
df_setor_pedidos['ticket_medio'] = df_setor_pedidos['faturamento']/df_setor_pedidos['qtd_pedidos']
df_setor_pedidos['ltv'] = df_setor_pedidos['ticket_medio'] * df_setor_pedidos['qtd_pedidos']/df_setor_pedidos['qtd_clientes']

# Exibir o DataFrame resultante
display(df_setor_pedidos)
```

```python
# Remover colunas intermediárias e alocar a um novo DataFrame
df_setores_ltv = df_setor_pedidos.drop(columns=['qtd_pedidos', 'qtd_clientes', 'faturamento'])
display(df_setores_ltv)
```

```python
# Fazer contagem de pedidos e clientes por setor considerando apenas recompradores
df_setor_pedidos_r = df_merge_recompradores[['setor_varejo', 'id_pedido', 'id_cliente_unica']].groupby('setor_varejo').nunique().reset_index()
df_setor_pedidos_r.rename(columns={'id_pedido': 'qtd_pedidos', 'id_cliente_unica': 'qtd_clientes'}, inplace=True)

# Mesclar o DataFrame de setores para obter o faturamento obtido dos recompradores
df_setor_pedidos_r = df_setor_pedidos_r.merge(df_fat_setores_recompradores, left_on='setor_varejo', right_on='setor_varejo')

# Calcular o ticket médio e LTV para os recompradores
df_setor_pedidos_r['ticket_medio_recompra'] = df_setor_pedidos_r['faturamento']/df_setor_pedidos_r['qtd_pedidos']
df_setor_pedidos_r['ltv_recompra'] = df_setor_pedidos_r['ticket_medio_recompra'] * df_setor_pedidos_r['qtd_pedidos']/df_setor_pedidos_r['qtd_clientes']

# Exibir o DataFrame resultante
df_setor_pedidos_r
```

```python
# Remover colunas intermediárias e alocar a um novo DataFrame
df_setores_recompradores_ltv = df_setor_pedidos_r.drop(columns=['qtd_pedidos', 'qtd_clientes', 'faturamento'])
display(df_setores_recompradores_ltv)
```

## Comparativo entre o crescimento dos setores de varejo na Olist e no mercado geral

```python
# Obter o faturamento de 2016 para os setores de varejo na Olist
df_set_2016 = df_merge2[df_merge2['ano_aprov_compra'] == 2016]
df_set_2016 = df_set_2016.groupby('setor_varejo')[['setor_varejo', 'preco']].sum()
df_set_2016.rename(columns={'preco': 'fat_2016_olist'}, inplace=True)

# Obter o faturamento de 2017 para os setores de varejo na Olist
df_set_2017 = df_merge2[df_merge2['ano_aprov_compra'] == 2017]
df_set_2017 = df_set_2017.groupby('setor_varejo')[['setor_varejo', 'preco']].sum()
df_set_2017.rename(columns={'preco': 'fat_2017_olist'}, inplace=True)

# Obter o faturamento de 2018 para os setores de varejo na Olist
df_set_2018 = df_merge2[df_merge2['ano_aprov_compra'] == 2018]
df_set_2018 = df_set_2018.groupby('setor_varejo')[['setor_varejo', 'preco']].sum()
df_set_2018.rename(columns={'preco': 'fat_2018_olist'}, inplace=True)

# Unir os DataFrames e exibir o resultado
df_setores_olist = df_set_2016.merge(df_set_2017, on='setor_varejo')
df_setores_olist = df_setores_olist.merge(df_set_2018, on='setor_varejo')
display(df_setores_olist)
```

Notamos que o faturamento de 2016 é muito baixo, possivelmente devido ao período (os dados mais antigos são do final de 2016), então vamos nos focar no crescimento entre 2017 e 2018 para uma comparação mais equilibrada.

```python
# Calcular os crescimentos de 2016 a 2017, 2017 a 2018 e 2016 a 2018
df_setores_olist['cresc_2017_olist'] = (df_setores_olist['fat_2017_olist'] - df_setores_olist['fat_2016_olist'])/df_setores_olist['fat_2016_olist']
df_setores_olist['cresc_2018_olist'] = (df_setores_olist['fat_2018_olist'] - df_setores_olist['fat_2017_olist'])/df_setores_olist['fat_2017_olist']
df_setores_olist['cresc_tot_olist'] = (df_setores_olist['fat_2018_olist'] - df_setores_olist['fat_2016_olist'])/df_setores_olist['fat_2016_olist']

# Remover colunas indesejadas e exibir o DataFrame resultante
df_setores_olist.drop(columns=['fat_2016_olist', 'fat_2017_olist', 'fat_2018_olist'], inplace=True)
display(df_setores_olist)
```

```python
# Mesclar com o DataFrame com os dados do varejo no mercado como um todo
df_setores_comp = df_setores_olist.merge(df_setores_varejo[['setor_varejo', 'fat_2016', 'fat_2017', 'fat_2018']], on='setor_varejo')
display(df_setores_comp)
```

```python
# Calcular os crescimentos de 2016 a 2017, 2017 a 2018 e 2016 a 2018 referentes
# ao mercado como um todo
df_setores_comp['cresc_2017'] = (df_setores_comp['fat_2017'] - df_setores_comp['fat_2016'])/df_setores_comp['fat_2016']
df_setores_comp['cresc_2018'] = (df_setores_comp['fat_2018'] - df_setores_comp['fat_2017'])/df_setores_comp['fat_2017']
df_setores_comp['cresc_tot'] = (df_setores_comp['fat_2018'] - df_setores_comp['fat_2016'])/df_setores_comp['fat_2016']

# Eliminar colunas indesejadas e exibir o DataFrame resultante
df_setores_comp.drop(columns=['fat_2016', 'fat_2017', 'fat_2018'], inplace=True)
display(df_setores_comp)
```

```python
# Exibir gráfico de colunas comparando o crescimento de 2017 a 2018 na Olist
# e no mercado geral
df_setores_comp.plot.bar(x='setor_varejo', y=['cresc_2018_olist', 'cresc_2018'])
plt.title('Crescimento (%) de 2017 a 2018 dos setores de varejo na Olist e no mercado geral')
plt.legend(['Crescimento do setor na Olist', 'Crescimento do setor no mercado geral'])
plt.xlabel('')
plt.show()
```
